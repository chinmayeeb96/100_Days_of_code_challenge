```python
import numpy as np
from scipy.stats import rankdata
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style ='whitegrid',palette="dark")

data_ = np.arange(1,30).reshape(-1,1)
data_[-1] = data_[-1] * 100
data_.reshape(-1)
```

```python
sns.lineplot(data=data_, marker="o")

plt.savefig('/home/chinmayee/my_machine_learning_journey/100_Days_of_code_challenge/figures/non_rank.png', dpi=300)
```

```python
rank_data = rankdata(data_)
sns.lineplot(data=rank_data, marker="^")

plt.savefig('/home/chinmayee/my_machine_learning_journey/100_Days_of_code_challenge/figures/rank.png', dpi=300)
```
