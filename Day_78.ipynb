## Let's generate some features from the datetime(InvoiceDate) feature

```python
#df['timestemp'] = pd.to_datetime(df['InvoiceDate'], format = '%m/%d/%Y %H:%M')
timestemp_dt= pd.to_datetime(df['InvoiceDate'], format = '%m/%d/%Y %H:%M').dt
#df['date'] = pd. to_datetime(df['InvoiceDate']). dt

#df['time'] = pd. to_datetime(df['InvoiceDate']). dt. time
```

```python
df['day'] = timestemp_dt.day
df['day_name'] = timestemp_dt.day_name()
df['weekOfyear'] = timestemp_dt.isocalendar().week 
df['dayofweek'] = timestemp_dt.dayofweek
df['dayofyear'] = timestemp_dt.dayofyear
df['quarter'] = timestemp_dt.quarter
df['month'] = timestemp_dt.month
df['month_name'] = timestemp_dt.month_name()
df.tail(15)
```

```python
df['is_month_start'] = timestemp_dt.is_month_start # Indicates whether the date is the first day of the month.
df['is_month_end'] = timestemp_dt.is_month_end # Indicates whether the date is the last day of the month.
df['is_quarter_start'] = timestemp_dt.is_quarter_start
df['is_quarter_end'] = timestemp_dt.is_quarter_end
df['is_year_start'] = timestemp_dt.is_year_start
df['is_year_end'] = timestemp_dt.is_year_end
df['is_leap_year'] = timestemp_dt.is_leap_year
df['daysinmonth'] = timestemp_dt.daysinmonth # The number of days in the month.
df.head(10)
```
