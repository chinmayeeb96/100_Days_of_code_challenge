# <b> Problem 8 : Create a new column for the updated price. This updated price will be calculated on top of the given price - if the engine is in front, price will be same else if the engine is in rear, price will be doubled </b>


```python
automobile_df['engine-location'].unique()
```

```python
def revised_price(engine_location,price):
  if engine_location == 'rear':
    updated_price = (2*price)
  if engine_location == 'front' :
    updated_price = price
  return updated_price
```

```python
automobile_df['engine-location'].dtype
```

```python
automobile_df.rename(columns={'engine-location' : 'engine_location'}, inplace = True)
automobile_df.head()
```

```python
# Your code here. Use apply and lambda
automobile_df['updated_price'] = automobile_df.apply(lambda x : revised_price(x['engine_location'],x['price']), axis=1)


```

# <b> Problem 10 : Create a new column which stores the number of doors in a car as integers. 

```python
# Your code here. Use apply

automobile_df.info()
```

```python
automobile_df.rename(columns={'num-of-doors' : 'num_of_doors'}, inplace = True)
automobile_df.head()
```

```python
automobile_df['num_of_doors'].unique()
```

```python
def str_to_inte(num_of_doors):
    
    if num_of_doors == 'four':
        num = 4
    elif num_of_doors == 'two':
        num = 2
    else:
        num = 0
    return num
```

```python
automobile_df['doors'] = automobile_df.apply(lambda x : str_to_inte(x['num_of_doors']), axis=1)
automobile_df['doors'].dtype
automobile_df.head(2)
```
