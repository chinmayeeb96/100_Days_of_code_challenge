```python
from sklearn.datasets import make_blobs


X, y = make_blobs(n_samples=150, 
                  n_features=2, 
                  centers=3, 
                  cluster_std=0.5, 
                  shuffle=True, 
                  random_state=0)
print(X)
```

```python
import matplotlib.pyplot as plt


plt.scatter(X[:, 0], X[:, 1], 
            c='g', marker=r'$\clubsuit$', s=100)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Unlabeled dataset')

plt.grid()
plt.tight_layout()
plt.savefig('/home/chinmayee/my_machine_learning_journey/100_Days_of_code_challenge/figures/clustering.png', dpi=1000)
plt.show()
```
