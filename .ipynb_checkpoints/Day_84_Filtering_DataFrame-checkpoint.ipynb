```python
# To choose certain columns simultaneously with the above filtering

df.loc[
    df.radius_mean > 25,
    ['radius_mean','texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean']
]
```

### Using bitwise operators

```python
# AND operator

df.loc[
    (df.diagnosis == 'M') & (df.radius_mean > 25),
    ['diagnosis','radius_mean','texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean']
    
]
```

```python
# OR operator

df.loc[
    (df.diagnosis == 'M') | (df.radius_mean > 25),
    ['diagnosis','radius_mean','texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean']
    
]
```

```python
# using between

df.loc[
    (df.diagnosis == 'M') | (df.radius_mean.between(24, 28) ),
    ['diagnosis','radius_mean','texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean']
    
]
```

```python
# isin() method. It tells if a particular value exists in a particular column
df.loc[
    df.radius_mean.isin([19.69]),
    ['diagnosis','radius_mean','texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean']
    
]
```

#### Index of the maximum and minimum value of a given column

```python
[df.radius_mean.idxmax(), df.radius_mean.idxmin()]
```

```python
df.loc[
    [df.radius_mean.idxmax(), df.radius_mean.idxmin()],
    ['diagnosis','radius_mean','texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean']
    
]
```

```python
# grabing columns of a dataframe by passing a list to items:
df.filter(items=['diagnosis', 'radius_mean']).head()
```

```python
# use regular expressions; here, we select any columns that start with t:
df.filter(regex = r'^c').head()
```
