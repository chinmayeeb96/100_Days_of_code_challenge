### DBSCAN: Density-based spatial clustering of applications with noise 


```python
from sklearn.datasets import make_moons
import matplotlib.pyplot as plt

X,y = make_moons(n_samples=200, noise=0.05, random_state=0)
plt.scatter(X[:,0],X[:,1])

plt.xlabel('feature_1')
plt.ylabel('feature_2')
plt.tight_layout()
plt.show()
```

```python
from sklearn.cluster import DBSCAN

db = DBSCAN(eps=0.2, min_samples=5, metric='euclidean')
y_db = db.fit_predict(X)

plt.scatter(X[y_db==0,0], X[y_db==0,1],c='lightgreen', marker='o',s=40,label='Cluster_1')
plt.scatter(X[y_db==1,0], X[y_db==1,1],c='purple', marker='s',s=40,label='Cluster_2')

plt.xlabel('feature_1')
plt.ylabel('feature_2')

plt.legend()

plt.tight_layout()
plt.savefig('/home/chinmayee/my_machine_learning_journey/100_Days_of_code_challenge/figures/DBSCAN.png', dpi=300)
plt.show()


```
