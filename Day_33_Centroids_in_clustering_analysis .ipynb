```python
from sklearn.datasets import make_blobs


X, y = make_blobs(n_samples=150, 
                  n_features=2, 
                  centers=3, 
                  cluster_std=0.5, 
                  shuffle=True, 
                  random_state=0)
print(X)
```

```python
import matplotlib.pyplot as plt


plt.scatter(X[:, 0], X[:, 1], 
            c='g', marker=r'$\clubsuit$', s=100)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Unlabeled dataset')

plt.grid()
plt.tight_layout()
plt.savefig('/home/chinmayee/my_machine_learning_journey/100_Days_of_code_challenge/figures/clustering.png', dpi=1000)
plt.show()
```

```python
from sklearn.cluster import KMeans


km = KMeans(n_clusters=3, 
            init='random', 
            n_init=10, 
            max_iter=300,
            tol=1e-04,
            random_state=0)

y_km = km.fit_predict(X)
```

```python
plt.scatter(X[y_km == 0, 0],
            X[y_km == 0, 1],
            s=50, c='green',
            marker='s', edgecolor='black',
            label='Cluster 1')
plt.scatter(X[y_km == 1, 0],
            X[y_km == 1, 1],
            s=50, c='blue',
            marker='o', edgecolor='black',
            label='Cluster 2')
plt.scatter(X[y_km == 2, 0],
            X[y_km == 2, 1],
            s=50, c='brown',
            marker='v', edgecolor='black',
            label='Cluster 3')
plt.scatter(km.cluster_centers_[:, 0],
            km.cluster_centers_[:, 1],
            s=250, marker='*',
            c='red', edgecolor='black',
            label='Centroids')

plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

plt.legend(scatterpoints=1)
plt.grid()
plt.tight_layout()
plt.savefig('/home/chinmayee/my_machine_learning_journey/100_Days_of_code_challenge/figures/clustering_centroi.png', dpi=1000)
plt.show()
```
